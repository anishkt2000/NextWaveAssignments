```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import radians, sin, cos, sqrt, atan2
from datetime import datetime, timedelta

def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Earth's radius in kilometers
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance_km = R * c
    return distance_km * 0.539957  # Convert to nautical miles

# Sample data
data = [
    (1, 'SOSP', 43831.708333, 'Port A', 34.0522, -118.2437, '9434761', '6', None),
    (2, 'EOSP', 43831.791667, 'Port B', 36.7783, -119.4179, '9434761', '6', None),
    (3, 'SOSP', 43832.333333, 'Port B', 36.7783, -119.4179, '9434761', '6', None),
    (4, 'EOSP', 43832.583333, 'Port C', 37.7749, -122.4194, '9434761', '6', None),
    (5, 'SOSP', 43833.000000, 'Port C', 37.7749, -122.4194, '9434761', '6', None),
    (6, 'EOSP', 43833.250000, 'Port D', 32.7157, -117.1611, '9434761', '6', None)
]

df = pd.DataFrame(data, columns=['id', 'event', 'datetime', 'voyage_From', 'lat', 'lon', 'imo_num', 'voyage_Id', 'allocatedVoyageId'])

# Filter for specific vessel and voyage
df = df[(df['imo_num'] == '9434761') & (df['voyage_Id'] == '6') & (df['allocatedVoyageId'].isnull())]

# Convert Excel date to UTC datetime
df['utc_datetime'] = df['datetime'].apply(lambda x: datetime(1899, 12, 30) + timedelta(days=x))

# Sort by UTC datetime
df = df.sort_values('utc_datetime')

# Identify voyage stages
df['stage'] = (df['event'] == 'SOSP').cumsum()
df['stage_type'] = df['event'].map({'SOSP': 'Sea Passage', 'EOSP': 'Port Stay'})

# Calculate time differences
df['duration'] = df['utc_datetime'].diff()
df['duration_hours'] = df['duration'] / pd.Timedelta(hours=1)

# Calculate distances
df['distance_nautical_miles'] = df.apply(lambda row: haversine_distance(
    row['lat'], row['lon'],
    df.loc[df.index[df.index.get_loc(row.name) - 1], 'lat'] if row.name > df.index[0] else row['lat'],
    df.loc[df.index[df.index.get_loc(row.name) - 1], 'lon'] if row.name > df.index[0] else row['lon']
), axis=1)

# Calculate cumulative times and distances for each stage
df['cumulative_hours'] = df.groupby('stage')['duration_hours'].cumsum()
df['cumulative_distance'] = df.groupby('stage')['distance_nautical_miles'].cumsum()

# Calculate stage statistics
stage_stats = df.groupby('stage').agg({
    'stage_type': 'first',
    'voyage_From': lambda x: f"{x.iloc[0]} to {x.iloc[-1]}",
    'duration_hours': 'sum',
    'distance_nautical_miles': 'sum',
    'utc_datetime': ['min', 'max']
}).reset_index()

stage_stats.columns = ['stage', 'stage_type', 'route', 'duration_hours', 'distance_nautical_miles', 'start_time', 'end_time']

# Visualization
plt.figure(figsize=(12, 6))

for stage, group in df.groupby('stage'):
    color = 'blue' if group['stage_type'].iloc[0] == 'Sea Passage' else 'red'
    plt.plot(group['utc_datetime'], group['cumulative_hours'], color=color, label=f"Stage {stage}: {group['stage_type'].iloc[0]}")
    plt.scatter(group['utc_datetime'], group['cumulative_hours'], color=color)

for _, row in df.iterrows():
    plt.annotate(f"{row['event']} at {row['voyage_From']}\n{row['utc_datetime'].strftime('%Y-%m-%d %H:%M:%S')}", 
                 (row['utc_datetime'], row['cumulative_hours']),
                 xytext=(0, 10), textcoords='offset points', ha='center', va='bottom',
                 bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
                 arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))

plt.xlabel('Date and Time (UTC)')
plt.ylabel('Cumulative Hours')
plt.title('Voyage Timeline for IMO 9434761, Voyage 6')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Print results
print("Voyage Events:")
print(df.to_string(index=False))
print("\nVoyage Stage Statistics:")
print(stage_stats.to_string(index=False))
```
